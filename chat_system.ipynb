###  Create a Chat System using Object-Oriented Programming (OOP) concepts. (Marks 4) You need to create the following classes:
    User
    Message
    ChatRoom
The system should implement the following functionalities:
####
    Sending messages
    Viewing chat history
    User joining and leaving the chat room

```python
import datetime # for tracking the user joining and leaving the chat


class User:
    def __init__(self, username):
        self.username = username

class Message:
    def __init__(self, sender, text):
        self.sender = sender
        self.text = text
        self.time = datetime.datetime.now().strftime("%H:%M")


class ChatRoom:
    def __init__(self, username):
        self.username = username
        # Dictionary so that we keep the history of user for easy tracking
        self.active_users = {} 
        self.chat_history = []

    def join_room(self, user_obj):
        
        self.active_users[user_obj.username] = user_obj
        print(user_obj.username + " has entered the chat.")

    def leave_room(self, user_obj):
        if user_obj.username in self.active_users:
            del self.active_users[user_obj.username]
            print(user_obj.username + " has left the chat.")

    def send_msg(self, user_obj, message_text):
        
        if user_obj.username in self.active_users:
            new_msg = Message(user_obj.username, message_text)
            self.chat_history.append(new_msg)
        else:
            print("Error: User not in room!")

    def show_all_messages(self):
        print("\n--- " + self.username + " History ---")
        for m in self.chat_history:
            print("[" + m.time + "] " + m.sender + ": " + m.text)

```

```python

# OOP Based Chat System

student_1 = User("Fizza")
student_2 = User("Tooba")

# my ChatRoom Name
my_room = ChatRoom("AI Group")

# Fully functional chating 

my_room.join_room(student_1)
my_room.join_room(student_2)

my_room.send_msg(student_1, "AOA, assignment complete ho gayi?")
my_room.send_msg(student_2, "Walaikum Assalam, ji bas chat system logic implement krlun phir testing kr k upload krdungi.")

my_room.show_all_messages()

my_room.leave_room(student_1)
```
