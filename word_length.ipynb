### Given a list of words: words = ["apple", "banana", "kiwi", "cherry", "mango"] (Marks 3) Create a dictionary that maps each word to its corresponding length.

Example Output: ({"apple": 5, "banana": 6, "kiwi": 4, "cherry": 6, "mango": 5})


```python
words = ["apple", "banana", "kiwi", "cherry", "mango"] 
word_length = [len(w) for w in words]

output = dict(zip(words,word_length))

print(f"Input: {words}")
print(f"Output: {output}")
```
