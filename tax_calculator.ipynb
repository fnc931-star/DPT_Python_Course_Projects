### Write a program that takes salary as input. Using conditional statements, calculate the final tax based on the following rules: (Marks 3)
 	If the salary is less than 30,000→Tax rate is 5%
	If the salary is between 30,000 and 70,000→Tax rate is 15%
	If the salary is greater than 70,000→Tax rate is 25%


```python

def tax_calculator(salary):
    if  0 < salary < 30000:
        tax = 5
        tax_rate = (5/100) * salary
        return tax, tax_rate

    elif  30000 <= salary <= 70000:
        tax = 15
        tax_rate = (15/100) * salary
        return tax, tax_rate

    elif  salary > 70000:
        tax = 25
        tax_rate = (25/100) * salary
        return tax, tax_rate

    

salary = int(input("Enter Your Salary: "))

if salary > 0:
    tax, result = tax_calculator(salary)
    print(f"The tax on {salary} is {tax}% - in price {result}")
else:
    print("Ivalid input!")


```
